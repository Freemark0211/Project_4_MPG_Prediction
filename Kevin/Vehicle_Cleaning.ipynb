```python
# function to automate reductions

def clean_up(df, column, replace):
    
    for c in replace:
        df.loc[vr[column].str.contains(c), column] = c
        
```

```python
# reduce VClasses

vr = sorted.copy()
class_search_values = ["Minivan", "Small Pickup Trucks", "Special Purpose Vehicle", "Sport Utility Vehicle",\
                      "Standard Pickup Trucks","Vans", "Small Sport Utility Vehicle", "Standard Sport Utility Vehicle"]

# for c in class_search_values:
#     vr.loc[vr["VClass"].str.contains(c), "VClass"] = c
clean_up(vr, "VClass", class_search_values)    
    
vr.loc[vr["VClass"].str.contains("Midsize Station"), "VClass"] = "Midsize-Large Station Wagons"
    
vr["VClass"].value_counts()
```

```python
# reduce to 4, front or rear drive

vr.loc[vr["drive"].str.contains("2-Wheel"), "drive"] = "Front-Wheel Drive"
vr.loc[vr["drive"].str.contains("All-Wheel"), "drive"] = "4-Wheel Drive"
vr.loc[vr["drive"].str.contains("Part-time"), "drive"] = "4-Wheel Drive"
drive_counts = vr.drive.value_counts()
drive_counts.sort_index()
```

```python
# reduce to automatic or manual

AorM = ["Automatic", "Manual"]

clean_up(vr, "trany", AorM )

trany_counts = vr.trany.value_counts()
trany_counts.sort_index()
```

```python
fuel_count = vr.fuelType.value_counts()
fuel_count
```

```python
# remove non-gasoline fuels

vr = vr[(vr.fuelType == "Regular") | (vr.fuelType == "Premium")]
vr
```

```python
vr = vr.drop("fuelType", axis=1)
```

```python
vr = vr.rename(columns={"comb08": "MPG(comb)", "VClass": "vehicle_class","displ":"displ (L)"})
vr
```
